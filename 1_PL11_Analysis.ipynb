```python
df
```

```python
for i in range(99):
  df['Capacity'][pl09.iloc[i][1]-1] = pl09['Capacity'][i]

for i in range(94):

  if (df['Capacity'][pl19.iloc[i][1]-1]==None):
    df['Capacity'][pl19.iloc[i][1]-1] = pl19['Capacity'][i]
  else:
    df['Capacity'][pl19.iloc[i][1]-1] = np.mean([pl19['Capacity'][i],df['Capacity'][pl19.iloc[i][1]-1]])
```

```python
for i in range(528):
  df['CyclesLog'][i] = math.log10(i+1)
```

```python
df.info()
```

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='flare_r')
plt.gcf().set_size_inches(10,6)
```

```python
#based on the results, we have lots of missing data. One way to fill this data is to estimate the data based on its above and below non null data
for i in range(528):
  if (df['Capacity'][i]==None):
    j = i+1
    while j<=528:
      if (df['Capacity'][j]!=None):
        break
      j+=1
    df['Capacity'][i] = np.mean([df['Capacity'][i-1],df['Capacity'][j]])
```

```python
df['Capacity'] = pd.to_numeric(df['Capacity'])
```

Battery Capacity

```python
from sklearn.model_selection import train_test_split
import tensorflow as tf
from keras.layers import Dense, Dropout, Activation
from keras import callbacks
```

```python
#Split the data into feature and label subsets
x = pl11['PerEn3']
y = pl11['Capacity']
y = np.array([y]).reshape(-1,1)

#Split the data into train and test subsets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.1, random_state = 42)
```

```python
#ANN structure
ann = tf.keras.models.Sequential()
ann.add(tf.keras.layers.Dense(units=5,activation='relu'))
ann.add(tf.keras.layers.Dense(units=5,activation='relu'))
ann.add(tf.keras.layers.Dense(units=1))
ann.compile(optimizer = 'adam' , loss = 'mean_squared_error' , metrics = ['mean_squared_error'])
history = ann.fit(X_train,y_train,batch_size=32,epochs=3000, validation_data=(X_test, y_test))
```

```python
history_df = pd.DataFrame(history.history)

plt.plot(history_df['mean_squared_error'], label='Training MSE')
plt.plot(history_df['val_mean_squared_error'], label='Validation MSE')

plt.title('Training and Validation Mean Squared Error')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
history_df = pd.DataFrame(history.history)

plt.plot(history_df.loc[:, ['loss']], label='Training loss')
plt.plot(history_df.loc[:, ['val_loss']], label='Validation loss')
plt.title('Training and Validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend(loc="best")

plt.show()
```

```python
y_pred = ann.predict(X_test)
np.concatenate((y_pred.reshape(14,1),y_test.reshape(14,1)),1)
```

```python
plt.plot(y_test, label='Test Data')
plt.plot(y_pred, color='red', label='Predicted Data')
plt.legend()
plt.show()
```

```python
from sklearn import metrics
print('MAE : ' , metrics.mean_absolute_error(y_test,y_pred))
print('MSE : ' , metrics.mean_squared_error(y_test,y_pred))
print('RMSE : ' , np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

Battery Charge Condition

```python
y = y/pl11['Capacity'][0]
for i in range(len(pl11)):
  if y[i]<0.9:
    y[i]=0
  else:
    y[i]=1 

X_train, X_test, y_train, y_test = train_test_split( x, y, test_size=0.1, random_state=101)
```

```python
ann = tf.keras.models.Sequential()
ann.add(tf.keras.layers.Dense(units=10, activation='relu'))
ann.add(tf.keras.layers.Dense(units=20, activation='relu'))
ann.add(tf.keras.layers.Dense(units=30, activation='relu'))
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

```python
ann.compile(optimizer = 'adam' , loss = 'binary_crossentropy' , metrics = ['binary_crossentropy'])
```

```python
ann.fit(X_train,y_train,batch_size=8,epochs=100, validation_data=(X_test, y_test))
```

```python
y_pred = ann.predict(X_test)
plt.plot(y_test, '.', label='Test Data')
plt.plot(y_pred, label='Predicted Data')
plt.legend()
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
y_pred = ann.predict(X_test)
y_pred = (y_pred > 0.5)
print(confusion_matrix(y_test,y_pred))
```
